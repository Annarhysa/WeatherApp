```python
import requests
import pandas as pd 
import json
import datetime

city = input()
date = input()

URL ='https://api.oikolab.com/weather'
API_KEY= 'c61cfceb2f1243cbb62098367e2dee16'

resp = requests.get(URL, 
                    params = {'param':['temperature'], 
                           'start': date, 
                           'end': date, 
                           'location': city, 
                           'api-key': API_KEY, 
                           'freq': 'D'})
w = resp.json()['data']
```

```python
new_w=json.loads(w)
new_w
```

```python
df = pd.DataFrame(index = pd.to_datetime(new_w['index'], unit = 's'),
                  data = new_w['data'], columns = new_w['columns'])
temp = int(df.iloc[0,4])
print("Temparature for", city, "on", date, "=", temp, "C")
```
