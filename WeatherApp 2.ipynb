```python
import requests
import pandas as pd 
import json
import datetime

def get_temperature_variance(city, start_date, end_date):
  URL ='https://api.oikolab.com/weather'
  API_KEY= 'c61cfceb2f1243cbb62098367e2dee16'
  resp = requests.get(URL, 
                    params = {'param':['temperature'], 
                           'start': start_date, 
                           'end': end_date, 
                           'location': city, 
                           'api-key': API_KEY, 
                           'freq': 'D'})
  w = resp.json()['data']
  new_w=json.loads(w)
  df = pd.DataFrame(index = pd.to_datetime(new_w['index'], unit = 's'),
                  data = new_w['data'], columns = new_w['columns'])
  variance = df['temperature (degC)'].var()
  return variance
```

```python
city1 = input()
city2 = input()
start_date = input()

start_date = start_date.split('-')
obj1 = datetime.datetime(int(start_date[0]),int(start_date[1]),int(start_date[2]))
#obj.strftime('%y-%m-%d') from string to time
obj2 = obj1+datetime.timedelta(days=7) #to get end date

start_date = obj1.strftime('%y-%m-%d')
end_date = obj2.strftime('%y-%m-%d')

var1 = get_temperature_variance(city1, start_date, end_date)
var2 = get_temperature_variance(city2, start_date, end_date)

if var1 < var2:
  print(f"We choose {city1} beacause of less temperature variance")
else:
  print(f"We choose {city2} because of less temperature variance")
```
